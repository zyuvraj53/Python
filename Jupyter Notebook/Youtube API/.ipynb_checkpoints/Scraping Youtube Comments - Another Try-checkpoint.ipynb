```python
os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"
```

```python
api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = api_key
```

```python
youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey = DEVELOPER_KEY)
```

```python
request = youtube.commentThreads().list(
    part="snippet, replies",
    order="relevance",
    videoId="qy4OmeTuNQ8",
    maxResults = 100
)
```

```python
response = request.execute()
```

```python
data = response['items'][0]['snippet']['topLevelComment']['snippet']['textOriginal']
```

```python
data
```

```python
response
```

```python
topLevelComments = []
for i in range(0,100):
    topLevelComments.append(response['items'][i]['snippet']['topLevelComment']['snippet']['textOriginal'])
```

```python
topLevelComments
```

```python
type(topLevelComments)
```

```python
textfile = open("topLevelComments.js", "w",encoding="utf-8")
textfile.write("{[")
for element in topLevelComments:
    textfile.write("`" + element + "`," + "\n")
textfile.write("]}")
textfile.close()
```
